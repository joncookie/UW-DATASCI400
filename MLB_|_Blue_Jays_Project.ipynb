```python
import statsapi
from datetime import date, datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
```

```python
team_code = 'tor'
team = statsapi.lookup_team(team_code)
team_id = team[0]['id']
team_name = team[0]['teamName']
```

```python
#Get the beginning and ending of the week
day = date.today()- timedelta(days=14)
sd = day - timedelta(days=day.weekday())
ed = sd + timedelta(days=6)
sched = statsapi.schedule(start_date=sd,end_date=ed,team=team_id)
print(sd)
print(ed)
print(sched)
```

```python
for i in range(len(sched)):
  #get the game id for each game
  game_id = sched[i]['game_id']
  #Lookup box score for each game
  boxscore = statsapi.boxscore_data(game_id)
  #Get the date of the game
  game_date = datetime.strptime(boxscore['gameId'][:10], '%Y/%m/%d').date()

  print(boxscore)
  #Get opponent, if away team is not team_name then opponent is away team 
  #else home team
  if boxscore['teamInfo']['away']['teamName'] != team_name :
    opponent = boxscore['teamInfo']['away']['teamName']
    home_away = 'home'
  else:
    opponent = boxscore['teamInfo']['home']['teamName']
    home_away = 'away'
  
  #Extract venue and attendance
  for i in boxscore['gameBoxInfo']:
    if i['label']=='Venue':
      stadium = i['value'][0:-1]
    if i['label']=='Att':
      attendance = int(i['value'][0:-1].replace(',',''))

  print(boxscore[home_away]['teamStats'])
  #Batting Stats
  print(boxscore[home_away]['teamStats']['batting'])
  runs_for = boxscore[home_away]['teamStats']['batting']['runs']

  b_doubles = boxscore[home_away]['teamStats']['batting']['doubles']
  b_triples = boxscore[home_away]['teamStats']['batting']['triples']
  b_homers = boxscore[home_away]['teamStats']['batting']['homeRuns']
  b_hits = boxscore[home_away]['teamStats']['batting']['hits']
  b_singles = b_hits - b_doubles - b_triples - b_homers
  b_base_on_balls = boxscore[home_away]['teamStats']['batting']['baseOnBalls']
  b_strike_outs = boxscore[home_away]['teamStats']['batting']['strikeOuts']

  #Pitching Stats
  print(boxscore[home_away]['teamStats']['pitching'])
  runs_against = boxscore[home_away]['teamStats']['pitching']['runs']

  p_doubles = boxscore[home_away]['teamStats']['pitching']['doubles']
  p_triples = boxscore[home_away]['teamStats']['pitching']['triples']
  p_homers = boxscore[home_away]['teamStats']['pitching']['homeRuns']
  p_hits = boxscore[home_away]['teamStats']['pitching']['hits']
  p_singles = p_hits - p_doubles - p_triples - p_homers
  p_base_on_balls = boxscore[home_away]['teamStats']['pitching']['baseOnBalls']
  p_strike_outs = boxscore[home_away]['teamStats']['pitching']['strikeOuts']

  print("game_date: ", game_date)
  print("opponent: ", opponent)
  print("stadium:", stadium)
  print("attendance: ", attendance)
  print("runs_for:",runs_for)
```
